```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

# Example data (x, y coordinates + labels)
data = [
    (0.2, 0.1, "Kitten"),
    (0.4, 0.6, "Kitten"),
    (0.5, 0.2, "Kitten"),
    (0.7, 0.9, "Adult")
]

# Separate the points by class
x_cat = [x for x, y, label in data if label == "Kitten"]
y_cat = [y for x, y, label in data if label == "Kitten"]

x_dog = [x for x, y, label in data if label == "Adult"]
y_dog = [y for x, y, label in data if label == "Adult"]

# Plot
plt.scatter(x_cat, y_cat, color="red", label="Kitten", s=10, marker="o")
plt.scatter(x_dog, y_dog, color="black", label="Adult", s=10, marker="o")

# ploting epochs
# X range
x_vals = np.linspace(0, 5, 100)

# Example weights & bias from each epoch
# Suppose equations are w1*x1 + w2*x2 + b = 0
epoch_lines = [
    {"w1": -0.5, "w2": -0.8, "b": 0, "label": "Epoch 1"},
    {"w1": -1.0, "w2": -1.6, "b": 0, "label": "Epoch 2"},
    {"w1": -1.1, "w2": -1.8, "b": 1, "label": "Epoch 3"},
    {"w1": -1.4, "w2": -2.1, "b": 1, "label": "Epoch 4"},
    {"w1": -1.0, "w2": -1.5, "b": 2, "label": "Epoch 5"},
]

for e in epoch_lines:
    w1, w2, b = e["w1"], e["w2"], e["b"]
    if w2 != 0:  # avoid division by zero
        y_vals = -(w1/w2) * x_vals - b/w2
        plt.plot(x_vals, y_vals, linestyle="--", label=e["label"])

plt.title("Kitten vs Adult Classification Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
