####  Finally we take only two features SepalWidthCm and SepalLengthCm

```python
data = data.drop(["PetalLengthCm", "PetalWidthCm"], axis=1) # axis=0 => horizontal  # axis=1 => vertical


```

```python
print(data.head())
print(data.info())
print(data.describe())
```

delete the id column becasue this is not important

```python
data = data.drop("Id", axis=1)
```

#### save the the data with three column(independent variable sepal width & length , dependent variable = Species)

```python
data.to_csv("../Iris_Species_kaggle/Iris_modified_with_two_features.csv", index=False)
```

#### now normalizing both features with in range (0,1) and add dummy variable with replacement of catagorical variable which is target

```python
data['SepalLengthCm'] = (data['SepalLengthCm'] - data['SepalLengthCm'].min()) / (data['SepalLengthCm'].max() - data['SepalLengthCm'].min())
data['SepalWidthCm'] = (data['SepalWidthCm'] - data['SepalWidthCm'].min()) / (data['SepalWidthCm'].max() - data['SepalWidthCm'].min())

print(data.head())
print(data.describe())
```

```python
data['Species'] = data['Species'].map({'Iris-setosa': 0, 'Iris-virginica' : 1})
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

#### Save Final data-set for single_neuron alogirithmm

```python
data.to_csv("../Iris_Species_kaggle/Iris_normalized_freature_with_dummy_var.csv", index=False)
```
